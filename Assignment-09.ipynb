```python
#q2)
def print_exception_hierarchy(exception_class, indent=0):
    print(' ' * indent + exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass, indent + 4)

# Start with the base Exception class
print_exception_hierarchy(Exception)
```

```python
#q3)
#The ArithmeticError class is a base class for arithmetic-related exceptions in Python. 
#It serves as a superclass for various specific arithmetic-related exception classes.
#Some errors that are defined in the ArithmeticError class include:

#FloatingPointError: This exception is raised when a floating-point calculation fails. 
#It encompasses errors related to floating-point arithmetic, such as overflow, underflow, or invalid operations involving floating-point numbers.
try:
    result = float('inf') * 2  # Multiplying infinity by 2
except OverflowError as e:
    print(e)

#ZeroDivisionError: This exception is raised when a division or modulo operation is performed with zero as the divisor.
try:
    result = 10 / 0  # Dividing by zero
except ZeroDivisionError as e:
    print("Error:", str(e))
```

```python
#q4)
#The LookupError class in Python is a base class for exceptions that occur when an invalid lookup or index operation is performed on a sequence or 
#mapping
#KeyError: This exception is raised when a dictionary key or a set element is not found during a lookup operation.
my_dict = {"apple": 1, "banana": 2, "orange": 3}
try:
    value = my_dict["grape"]  # Accessing a non-existent key
except KeyError as e:
    print(e)
#IndexError: This exception is raised when a sequence (such as a list, tuple, or string) is accessed with an invalid index value
my_list = [1, 2, 3]
try:
    value = my_list[5]  # Accessing an element with an invalid index
except IndexError as e:
    print("Error:", str(e))
```

```python
#q5)
#ImportError: This exception is raised when an import statement fails to locate and import a module or when 
#there is an error during the import process.
try:
    import non_existent_module  # Trying to import a non-existent module
except ImportError as e:
    print("Error:", str(e))
#ModuleNotFoundError: This exception is a subclass of ImportError and is specifically raised when the specified module or package cannot 
#be found during the import process.
try:
    from non_existent_package import module_name  # Trying to import a non-existent module or package
except ModuleNotFoundError as e:
    print("Error:", str(e))
```
